```python
# prompt: add a legend,labels and title

import pandas as pd
import matplotlib.pyplot as plt

# Assuming the dataset is a CSV file named 'data.csv'
df = pd.read_csv('/content/anage.csv')

# Filter the DataFrame to include rows with non-null 'Body mass (g)' and 'Maximum longevity (yrs)' values
filtered_df = df[df['Body mass (g)'].notnull() & df['Maximum longevity (yrs)'].notnull()]

# Assuming 'Class' is the column you want to group by
# Replace 'Class' with the actual column name if it's different
groups = filtered_df.groupby('Class')

# Create a scatter plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed

for name, group in groups:
    plt.scatter(group['Body mass (g)'], group['Maximum longevity (yrs)'], label=name)

# Add labels and title
plt.xlabel('Body Mass (g)')
plt.ylabel('Maximum Longevity (yrs)')
plt.title('Correlation between Body Mass and Maximum Longevity by Class') # Added title

# Add a legend
plt.legend() # Added legend

# Display the plot
plt.grid(True)
plt.show()

```

```python
# prompt: use log scale for both x-axis and y-axis

import pandas as pd
import matplotlib.pyplot as plt

# Assuming the dataset is a CSV file named 'data.csv'
df = pd.read_csv('/content/anage.csv')

# Filter the DataFrame to include rows with non-null 'Body mass (g)' and 'Maximum longevity (yrs)' values
filtered_df = df[df['Body mass (g)'].notnull() & df['Maximum longevity (yrs)'].notnull()]

# Assuming 'Class' is the column you want to group by
# Replace 'Class' with the actual column name if it's different
groups = filtered_df.groupby('Class')

# Create a scatter plot with log scales
plt.figure(figsize=(10, 6))  # Adjust figure size as needed

for name, group in groups:
    plt.scatter(group['Body mass (g)'], group['Maximum longevity (yrs)'], label=name)

# Add labels and title
plt.xlabel('Body Mass (g)')
plt.ylabel('Maximum Longevity (yrs)')
plt.title('Correlation between Body Mass and Maximum Longevity by Class')

# Set log scales for both axes
plt.xscale('log')
plt.yscale('log')

# Add a legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
